```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
Landfil = pd.read_csv('Landfill_Site_airnote.csv')
```

```python
Landfil
```

```python
Landfil=Landfil.drop(['aqi','csecs','c00_30','c00_50','c02_50','motion','c01_00','voltage','csamples','humidity','pressure','temperature','lon','lat','location','device_uid','aqi_algorithm','serial_number','heatIndex','sensor','aqi_level','pm01_0cf1','pm02_5cf1','pm10_0cf1','pm01_0_rstd', 'pm02_5_rstd','pm10_0_rstd'],axis=1)
Landfil
```

```python
Landfil.captured=pd.to_datetime(Landfil.captured)
```

```python
Landfil
```

```python
Landfill=Landfil.set_index('captured')
Landfill
```

```python
Land=Landfill.rename({'captured':'Time','pm01_0':'$PM_{1}$','pm02_5':'$PM_{2.5}$','pm10_0':'$PM_{10}$'}, axis=1)
```

```python
Land
```

```python
Land.rename({'captured':'Time'}, axis=index)
```

```python
Landfill_1=Land.resample('3H').mean()
```

```python
Landfill_1
```

```python
Landfill_1.plot()
plt.xlabel('Days')
plt.ylabel('Pollution levels [$\mu g/m^3$ ]')
plt.axhline(y=15,color='orange',ls='--', label= 'WHO Threshold')
plt.axhline(y=35,color='black', ls= ':',label= 'Ghana Standard')
plt.savefig('Landfillsite.jpeg')
plt.legend(loc='upper left')
```
