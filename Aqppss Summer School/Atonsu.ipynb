```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
Atonsu = pd.read_csv('Atonsu_High_School_Junction.csv',skiprows=9, names=['ObjectID','Timestamp','Latitude','Longitude','Temperature','$PM_1$','$PM_{10}$','$PM_{2.5}$','Humidity'])
Abattoir =pd.read_csv('Kumasi_Abbattoir.csv')
Landfil = pd.read_csv('Landfill_Site_airnote.csv')
```

```python
Atonsu
```

```python
# Atonsu=Atonsu.drop('ObjectID',axis=1)
Atonsu.Timestamp=pd.to_datetime(Atonsu.Timestamp)
#At_index=Atonsu.set_index(Atonsu_timestamp)
At_index=Atonsu.set_index(Atonsu.Timestamp)
```

```python
Atonsu_1=At_index.drop('Timestamp',axis=1)
```

```python
type(Atonsu_1.index.values)
```

```python
Atonsu_2=Atonsu_1.resample('10T').mean()
```

```python
Atonsu_2.shape
```

```python
Atonsu_2[['$PM_1$','$PM_{2.5}$','$PM_{10}$']].plot(marker='*')
plt.axhline(y=15,color='orange',ls='--')
plt.axhline(y=35,color='black')
# plt.axhline(y=45,color='g',ls='--')
# plt.axhline(y=30,color='orange',ls='--')
plt.savefig('PM_Concentration.jpeg')
plt.ylabel('PM Concentration ($\mu$g/m$^3$)',fontsize=15)
plt.title('Air Quality at Atonsu High School Junction')
plt.savefig('Atonsu High School Junction.jpeg')
```

```python
(Atonsu_2[['$PM_1$','$PM_{2.5}$','$PM_{10}$']].cumsum()/1).plot(marker='*')
plt.axhline(y=15,color='orange',ls='--', label= 'WHO threshold')
plt.axhline(y=35,color='red',ls=':', label='Ghana Standard')
plt.ylabel('PM Concentration ($\mu$g/m$^3$)',fontsize=15)
plt.xlabel('Time')
plt.savefig('PM_C.jpeg')
plt.legend(loc='upper left')
```

```python
Atonsu_2.plot.scatter(x='Longitude',y='Latitude',c='$PM_{2.5}$',cmap='magma')
plt.ylim(6.6580,6.660)
plt.xlim(-1.597,-1.59)
```
