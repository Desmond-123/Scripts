```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import odeint
```

```python
def dU_dt(U,t):
    return [-alpha*U[0]*U[1], alpha*U[0]*U[1]-mu*U[1], mu*U[1]]
U0=[1500,1,0]

alpha, mu= 0.005, 0.1
ts = np.arange(0,60,0.01)
Us= odeint(dU_dt,U0,ts)
```

```python
S,I,R = Us[:,0],Us[:,1],Us[:,2]
```

```python
fig = plt.subplots(figsize=(12,10))
plt.plot(ts,S, linewidth=2.5, label='Susceptible')
plt.plot(ts,I, linewidth=2.5, label='Infected')
plt.plot(ts,R, linewidth=2.5, label='Recovered')
plt.legend(fontsize=15)
plt.show()
```
